```python
import pandas as pd

import sys
sys.path.append('../')

from ift6758.data import get_player_stats
```

```python
df = get_player_stats(2018, 'goalies')

df.head()
```

```python
# Sort the goalies by their save percentage (â€˜SV%â€™)

df1 = df.sort_values('SV%', ascending=False).head(20)
df1[['Player','GP','SA','SV','SV%']]
```

```python
#plot the top 10 goalies by SV% and their number of games played

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
sns.set(rc={'figure.figsize':(9,6)})

x1 = df1['SV%'].astype(float)
y1 = df1['Player']

plt.xlim(0.9, 1)

p1 = sns.barplot(x1, y1, orient= "h")

p1.set_title("Ranking Goalies - Save Percentage", fontsize=16, weight='bold')
p1.set(
    xlabel='Save Percentage (SV%)', 
    ylabel='Goalie name')



plt.show()


x2 = df1['GP'].astype(float)
y2 = df1['Player']

plt.xticks(rotation=90)
plt.axvline(41)

p2 = sns.barplot(x2, y2, orient= "h")

p2.set_title("# Games Top 20 Goalies (Save Percentage)", fontsize=16, weight='bold')
p2.set(
    xlabel='# Games', 
    ylabel='Goalie name')

plt.show()
```

```python
# Sort the goalies by their save percentage (â€˜SV%â€™), 
# considering only the goalies that played at least half of the games of the regular season (41 games)

df2 = df.sort_values('SV%', ascending=False)[df['GP'].astype(float)>40].head(20)
df2[['Player','GP','SA','SV','SV%']]
```

```python
sns.set_style('darkgrid')
sns.set(rc={'figure.figsize':(9,6)})

y1 = df2['Player']
x1 = df2['SV%'].astype(float)
plt.xlim(0.9, 0.95)



p1 = sns.barplot(x1, y1, orient= "h")

p1.set_title("Ranking Goalies - Save Percentage - At least 41 games", fontsize=16, weight='bold')
p1.set(
    xlabel='Save Percentage (SV%)', 
    ylabel='Goalie name')

plt.show()

x2 = df2['GP'].astype(float)
y2 = df2['Player']

plt.xticks(rotation=90)
plt.axvline(41)

p2 = sns.barplot(x2, y2, orient= "h")

p2.set_title("# Games Top 20 Goalies (Save Percentage - At least 41 games)", fontsize=16, weight='bold')
p2.set(
    xlabel='# Games', 
    ylabel='Goalie name')

plt.show()
```
