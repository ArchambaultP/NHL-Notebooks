```python
import pandas as pd

import sys
sys.path.append('../')

from ift6758.data import get_player_stats
```

```python
df = get_player_stats(2018, 'goalies')

df.head()
```

```python
# Sort the goalies by their save percentage (‘SV%’)

df1 = df.sort_values('SV%', ascending=False).head(10)
df1[['Player','GP','SA','SV','SV%']]
```

```python
#plot the top 10 goalies by SV% and their number of games played

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')

x1 = df1['SV%'].astype(float)
y1 = df1['Player']

plt.xlim(0.9, 1)

sns.barplot(x1, y1, orient= "h")

plt.show()


x2 = df1['GP'].astype(float)
y2 = df1['Player']

plt.xticks(rotation=90)
plt.axvline(41)

sns.barplot(x2, y2, orient= "h")

plt.show()
```

```python
# Sort the goalies by their save percentage (‘SV%’), 
# considering only the goalies that played at least half of the games of the regular season (41 games)

df2 = df.sort_values('SV%', ascending=False)[df['GP'].astype(float)>40].head(10)
df2[['Player','GP','SA','SV','SV%']]
```

```python
sns.set_style('darkgrid')

y1 = df2['Player']
x1 = df2['SV%'].astype(float)
plt.xlim(0.9, 0.95)
#plt.xticks(rotation=90)

sns.barplot(x1, y1, orient= "h")

plt.show()

x2 = df2['GP'].astype(float)
y2 = df2['Player']

plt.xticks(rotation=90)
plt.axvline(41)

sns.barplot(x2, y2, orient= "h")

plt.show()
```
